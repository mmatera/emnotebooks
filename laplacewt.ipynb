Resolución de la Ecuación de Laplace en un cable sometido a una diferencia de potencial armónica
=========================================


La idea es resolver la ecuación de Laplace en forma aproximada, para la transformada de Fourier del potencial respecto al tiempo, en presencia de un conductor cilíndrico:
$$
\nabla^2 \Phi = 0
$$
con $\Phi=\Phi(r,z)$ un potencial con simetría cilíndrica. Para esto, introducimos una celda elemental correspondiente a un elemento de volumen  $\Delta r \times \Delta z \times r \Delta \phi$, donde asumiremos además que $\Delta r=\Delta z=\Delta$. La ecuación de Laplace discretizada en esta geometría se expresa según

$$\Phi(r,z)= \frac{(1+\frac{\Delta}{2 r} )\Phi(r+\Delta,z)+(1-\frac{\Delta}{2 r} )\Phi(r-\Delta,z)+\Phi(r,z+\Delta)+\Phi(r,z-\Delta)}{4} + {\cal O}(\Delta^2)
$$

con la condición 
$\Phi(-\Delta,z)=\Phi(\Delta,z)$

Proponemos entonces que
$$
\Phi(\Delta i,\Delta j)=\Phi_{i,j}
$$

de manera que los $\Phi_{i,j}$ satisfacen las condiciones anteriores.

Si ahora consideramos regiones conductoras, la solución quedará determinada en términos de las condiciones de borde, impuestas sobre el conductor, y sobre los límites de la región a estudiar. En el caso de las condiciones de Dirichlet (potencial fijo) la condición se establece directamente fijando el valor del potencial en los correspondientes elementos $\Phi_{i,j}$

Condiciones tipo Neumann, o de discontinuidad en las derivadas se pueden imponer modificando la condición de actualización sobre los correspondientes elementos. Por ejemplo, si queremos imponer que la derivada radiar se anule en un elemento $i,j$, basta imponer
$$
\Phi_{i,j}=\Phi_{i-1,j}
$$

En la superficie de un conductor cilíndrico, de conductividad finita $\sigma$, la condición de conservación de la carga en su superficie lateral implica que
$$
\Sigma = -\frac{\sigma}{{\bf i}\omega}\check{r}\cdot\nabla \Phi
$$

con $\Sigma$ la densidad superficial de carga que, de acuerdo con la Ley de Gauss,
$$
\Sigma = \left.\varepsilon \check{r}\cdot\nabla \Phi\right|_{r=a-\epsilon}
-\left.\varepsilon \check{r}\cdot\nabla \Phi\right|_{r=a+\epsilon}
$$
por lo que
$$
\left.\check{r}\cdot\nabla \Phi\right|_{r=a+\epsilon}  =
\left.(\frac{\varepsilon_{int}}{\varepsilon_{ext} } +\frac{\sigma}{{\bf i}\omega\varepsilon_{ext}} )\check{r}\cdot\nabla \Phi\right|_{r=a-\epsilon} 
$$

 
Sobre las celdas correspondientes a la interface, podemos establecer esta condición de acuerdo con

$$
\Phi_{i,j}=\frac{\Phi_{(i+1),j}+ (\frac{\varepsilon_{int}}{\varepsilon_{ext}}+\frac{\sigma}{{\bf i}\omega \varepsilon_{ext}})\Phi_{(i-1),j}}{1+\frac{\varepsilon_{int}}{\varepsilon_{ext}}+\frac{\sigma}{{\bf i}\omega \varepsilon_{ext}}}
$$


Finalmente, para resolver el conjunto de relaciones entre los valores de una celda y sus vecinas, proponemos una configuración inicial en la que el potencial toma el valor especificado en las celdas sujetas a condiciones de Dirichlet, y en el resto toma valores cercanos a los reales. Luego, recorriendo las celdas actualizamos cada valor de acuerdo a las condiciones, e iteramos hasta la convergencia.




Calculo del campo en la superficie y densidad de carga
=============================================



Una vez resuelto el sistema, además de disponer de una descripción cualitativa del potencial (que mejorará al refinar el mallado) podemos estimar algunas cantidades de interés. Una de estas cantidades tiene que ver con las componentes del campo eléctrico. Estas pueden ser calculadas en diferencias finitas de acuerdo con

$$
\vec{E}_{r}\approx  \frac{\Phi_{i-1,j}-\Phi_{i+1,j}}{2\Delta}
$$
$$
\vec{E}_{z}\approx  \frac{\Phi_{i,j-1}-\Phi_{i,j+1}}{2\Delta}
$$

De manera análoga, podemos estimar la densidad superficial de carga en un punto de la interface  $r=a$
según
$$
\Sigma = \varepsilon_0 \frac{2\Phi_{i,j}-\Phi_{i+1,j}+\Phi_{i-1,j}}{\Delta}
$$
y sobre una interface con $z=\pm L/2$ según
$$
\Sigma = \varepsilon_0 \frac{2\Phi_{i,j}-\Phi_{i,j+1}+\Phi_{i,j-1}}{\Delta}
$$



Capacidad y momento dipolar. Parámetros adimensionales
================================================


A partir de la densidad superficial de carga podemos estimar la capacidad del conductor como el cociente entre la carga total separada y la diferencia de potencial aplicada.  Como el conductor es globalmente neutro, la carga separada puede expresarse como
$$
Q=\frac{1}{2}\iint |\Sigma| dS \approx \frac{\varepsilon_0 \Delta^2}{2} \sum_{i,j \in {\rm borde}}|\Sigma_{ij}|
$$
Luego,
$$
C=\frac{1}{2}\iint |\Sigma| dS \approx \frac{\varepsilon_0 \Delta^2}{2 (\Delta V)} \sum_{i,j \in {\rm borde}}|\Sigma_{ij}|
$$

De manera semejante, el momento dipolar inducido 
$$
\vec{p}=\frac{1}{2}\iint \vec{r}\Sigma dS \approx \frac{\varepsilon_0 \Delta^3}{2} \sum_{i,j \in {\rm borde}}j\Sigma_{ij}\check{e}_z
$$

Todas estas cantidades dependen explícitamente de las escalas de longitud y otros parámetros con dimensiones ($\sigma$, $\varepsilon_{int/ext}$, $\omega$). Al hacer simulaciones numéricas es conveniente expresar tanto los parámetros de entrada, como las cantidades calculadas como cantidades adimensionales.

Como entradas, conviene entonces trabajar con los cocientes
$$
\zeta=\frac{\varepsilon_{int}}{\varepsilon_{ext}}+\frac{\sigma}{{\bf i}\omega \varepsilon_{ext}}
$$
que define una permitividad relativa compleja entre los medios, y
$$
t=L/a
$$
la relación entre la longitud y el radio del alambre. 

El resto de los parámetros a introducir tienen que ver con la discretización del modelo en un recinto finito:
$n_z$ y $n_r$ son el número de celdas en que se divide la región a estudiar, y $L/\Delta=l*n_z$ el número de celdas que ocupa el alambre en la dirección del eje z. El límite del continuo se obtiene cuando $l n_z \rightarrow \infty$, mientras que el límite de "recinto infinito" (sin superficies limitadoras) se obtiene cuando $l\rightarrow 0$. Naturalmente, elegiremos $nz$ tan pequeño y $l$ tan grande como sea posible, de manera que las cantidades de interés se vuelvan independientes de estas cantidades.


Habiendo definido los parámetros que definen el problema, conviene introducir las cantidades
$$
c = \frac{C}{L}
$$
la capacidad por unidad  de longitud (medida en unidades de $\varepsilon_0$) y
$$
\vec{\cal P}=\frac{\vec{p}}{\Delta V \pi a^2}
$$
una "polarizabilidad efectiva", que indica cuanto vale el momento dipolar inducido en el volumen del alambre. Esta cantidad es importante para estimar la radiación dipolar eléctrica emitida por el alambre a grandes distancias.















```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

#Tamaño de la malla:
nz=400
nr=int(nz/2)
# dimensiones del alambre
l = .25       # lo bastante chica para tener suficiente "vacío" alrededor del cable.
a =  l     # relativo a l
##
DeltaV = 1.
epsint = None   # Asumimos que en el exterior, eps = eps0. Dentro, epsrel + \sigma/(I \omega)
omega = None     # la variable queda implícita en sigma. Sólo para fijar el límite omega=0
##

```

```python
jmax = None
jmin = None
boundary1 = None
boundary2 = None
boundary3 = None
pot = None



def inicializar():
    global pot, cable, exterior, boundary1, boundary2, boundary3, jmax, jmin
    pot = np.zeros((nr,nz),dtype=complex)
    # limites
    jmax = min(nz,int(.5*nz*(1+l)))-1
    jmin = max(0,int(.5*nz*(1.-l)))


    # Superficie lateral del cilindro
    boundary1 = [(int(nz*a),j) for j in  range(jmin,jmax+1)] 
    #Tapas
    boundary2 = [(i,jmin) for i in  range(0, int(nz*a))] 
    boundary3 = [(i,jmax) for i in  range(0, int(nz*a))] 


    cable = []
    exterior = []

    # Construimos listas de puntos adentro y afuera del cable:
    for i in range(1,nr-1):
        for j in range(1,nz-1):
            if (i,j) not in boundary1 and \
               (i,j) not in boundary2 and \
               (i,j) not in boundary3:
                if i <  int(a*nz) and   \
                        abs(2.*j/nz-1) < l:
                    cable.append((i,j))
                else:
                    exterior.append((i,j))

    
    # Relleno el exterior con la solución tipo alambre infinito/disco
    for p in exterior:
        if p[1]<jmin:
            cc = a*nz
            zz = (jmin-p[1]+.01)/cc
            cc = cc**2
            rrsq = p[0]**2/cc + zz**2
            zeta = np.sqrt(2)*zz/np.sqrt(np.sqrt((rrsq-1)**2+4*zz**2)-rrsq+1)
            pot[p] = .5*DeltaV*(1-2/np.pi*np.arctan(zeta))    
        elif p[1]>jmax:
            cc = a*nz
            zz = (p[1]-jmax+.01)/cc
            cc = cc**2
            rrsq = p[0]**2/cc + zz**2
            zeta = np.sqrt(2)*zz/np.sqrt(np.sqrt((rrsq-1)**2+4*zz**2)-rrsq+1)
            pot[p] = -.5*DeltaV*(1-2/np.pi*np.arctan(zeta))    
        else:
            val = -.5*DeltaV/l * (p[1]-.5*nz)/(jmax-jmin) 
            val = val * np.log( max(.5*l*nz/p[0],1 ) )/np.log( .5*l/(a+1/nz))
            pot[p] = val
    # Relleno el interior con la solución lineal
    for j in range(jmin,jmax):
        for i in range(0, int(nz*a)):
            val = -.5*DeltaV/l * (j-.5*nz)/(jmax-jmin) 
            pot[i,j] = val
    
    for p in boundary2:
        pot[p] = .5* DeltaV

    for p in boundary3:
        pot[p] = -.5* DeltaV




def update_pot(eps=None, sigma=None, omega=None):
    if pot is None or nr != len(pot) or nz != len(pot[0]):
        inicializar()
        
    if omega is not None:
        omega = omega
    if eps is not None:
        epsint = eps
        sigma = - epsint.imag
    elif sigma is not None:
        if omega != 0:
            epsint = 1 - sigma*1j/omega
        
    if epsint is None:
        return
    # Sobre el eje, condición de derivada nula
    for j in range(1,nz-1):
        if (0,j) in boundary2 or \
            (0,j) in boundary3:
            continue
        pot[0,j] = .25*(2.*pot[1,j]+pot[0,j+1]+pot[0,j-1])

    for p in cable:
        pot[p] =  .25*((1+.5/p[0])*pot[p[0]+1,p[1]]+(1-.5/p[0])*pot[p[0]-1,p[1]]+pot[p[0],p[1]+1]+pot[p[0],p[1]-1])

    for p in exterior:
        pot[p] =  .25*((1+.5/p[0])*pot[p[0]+1,p[1]]+(1-.5/p[0])*pot[p[0]-1,p[1]]+pot[p[0],p[1]+1]+pot[p[0],p[1]-1])

    if sigma == 0:
        for p in boundary1:
            pot[p] =  .25*((1+.5/p[0])*pot[p[0]+1,p[1]]+(1-.5/p[0])*pot[p[0]-1,p[1]]+pot[p[0],p[1]+1]+pot[p[0],p[1]-1])
    else:
        if omega == 0:
            for p in boundary1:
                # Derivada radial nula
                pot[p] = pot[p[0]-1,p[1]]
        else:
            
            for p in boundary1:
                pot[p] = (pot[p[0]+1,p[1]]+ epsint *pot[p[0]-1,p[1]])/(1+epsint)
                
#inicializar()


def carga_superficial():
    sigmas = []
    for p in boundary1[2:-2]:
        sigmas.append(2 * pot[p[0],p[1]]-pot[p[0]-1,p[1]]-pot[p[0]+1,p[1]] )
    return np.array(sigmas)


def capacidad_unidad_long():    
    sigmas = []
    for p in boundary2[:-2]:
        sigmas.append( (2*pot[p[0],p[1]]-pot[p[0],p[1]-1]-pot[p[0],p[1]+1]).real * p[0])
             
    for p in boundary3[:-2]:
        sigmas.append( (2*pot[p[0],p[1]]-pot[p[0],p[1]-1]-pot[p[0],p[1]+1]).real * p[0])
    
    qlateral = sum(abs(carga_superficial().real)) * a*nz
    q = sum([abs(s) for s in sigmas]) + qlateral
    return np.pi*q.real/DeltaV/(l *nz) 



def polarizabilidad():    
    sigmas = []
    for p in boundary2[:-1]:
        sigmas.append( (2*pot[p[0],p[1]]-pot[p[0],p[1]-1]-pot[p[0],p[1]+1]) * p[0])

    res = sum([s for s in sigmas])*(jmin)
    sigmas = []
    
    for p in boundary3[:-1]:
        sigmas.append( (2*pot[p[0],p[1]]-pot[p[0],p[1]-1]-pot[p[0],p[1]+1]) * p[0])

    res = res + sum([s for s in sigmas])*(jmax)
    res = res/(a*nz) + np.dot(carga_superficial(), np.array([p[1] for p in boundary1[2:-2]]))
    return res/(.5*a*nz*DeltaV)

 
```

```python
sigma = 0.
omega = 1.
l = .2
a = .2
nz = 50
nr = int(nz/2)

for j in range(10):
    for i in range(500):
        update_pot(eps=1.-.1*1j, sigma=sigma, omega=omega)
    print("   C/L: " ,capacidad_unidad_long(), "pF/m  P: " ,polarizabilidad(),"pC/m^2")
```

```python
########################### Graficos   ################################
fig,axs = plt.subplots(2,2,figsize=(12,8))

####          
ax = axs[0][0]

ax.contourf(pot.real)
bdry=np.array(boundary2+boundary1+boundary3)
ax.plot(bdry[:,1],bdry[:,0])
ax.set_title("Potencial en el plano r-z")
ax = axs[0][1]

ax.plot(pot[0,:].real,label="real eje")
ax.plot(pot[0,:].imag,label="imag eje")
ax.plot(pot[int(2*a*nz/3),:].real,label="real 2a/3")
ax.plot(pot[int(2*a*nz/3),:].imag,label="imag 2a/3")
ax.plot(pot[int(a*nz),:].real,label="real superficie")
ax.plot(pot[int(a*nz),:].imag,label="imag superficie")

ax.set_title("Potencial en el eje y en la superficie")
ax.legend()


ax = axs[1][0]


ax.plot(carga_superficial().real)
ax.set_title("Densidad superficial de carga en la cara lateral")



ax = axs[1][1]

sigmas = []
for p in boundary2[:-2]:
    sigmas.append( (2*pot[p[0],p[1]]-pot[p[0],p[1]-1]-pot[p[0],p[1]+1]).real)

ax.plot(np.array( boundary2[:-2])[:,0]/(a*nz), np.array(sigmas).real)
ax.set_title("Densidad superficial de carga en la tapa")
plt.show()
```
