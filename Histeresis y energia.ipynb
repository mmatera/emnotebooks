# Intercambios de energía entre el campo electromagnético y medios materiales


De la ley de fuerza de Lorentz se sigue que el trabajo que realizan los campos electromagnéticos sobre una distribución de cargas y corrientes por unidad de tiempo se expresa por
$$
W_{EM}=\int \vec{E}\cdot \vec{j}d{\cal V}
$$

Vía el teorema de Poynting, encontramos que esta cantidad satisface una relación semejante a la del teorema de trabajo y energía:
$$
W_{EM} + \int \vec{S}\cdot d\vec{\cal A}=-\frac{d}{dt} (U_E+U_B)
$$

con ${\cal V}$ el volumen en que se localiza el sistema considerado, ${\cal A}=\partial{\cal V}$,
$$
U_E=\int \frac{\varepsilon_0}{2}|\vec{E}|^2d{\cal V}
$$
$$
U_B=\int \frac{1}{2\mu_0}|\vec{B}|^2d{\cal V}
$$
la energía asociada a los campos eléctrico y magnético y
y 
$$
\vec{S}=\frac{\vec{E}\times \vec{B}}{\mu_0}
$$
el **vector de Poynting** que dá cuenta del flujo de energía hacia el exterior del sistema vía radiación.


En la expresión anterior, $\vec{j}$ involucra a todas las corrientes, tanto las que circulan por los conductores como las corrientes de magnetización inducidas en los medios materiales. En la práctica, las corrientes que podemos medir y controlar directamente son aquellas que circulan por conductores, las *corrientes de conducción*. 

Por esto, se introducen los campos auxiliares
$$
\vec{D}=\varepsilon_0 \vec{E}+\vec{\cal P}
$$
y 
$$
\vec{B}=\mu_0 (\vec{H}+\vec{\cal M})
$$
de manera que
$$
\nabla \cdot \vec{D}=\rho_c
$$
y
$$
\nabla \times \vec{H}=\vec{j}_c
$$

de forma que $\rho_p =\nabla\cdot\vec{\cal P}$ y $\vec{j}_M=\nabla\times \vec{\cal M}$ representan a las cargas y corrientes presentes en los medios materiales. Interesa entonces calcular el trabajo sobre estas cargas y corrientes que viene dado por

$$
W_c =\int \vec{E}\cdot \vec{j}_c d{\cal V}
$$

Siguiendo los pasos usados para mostrar el teorema de Poynting, encontramos ahora que

$$
W_c +\int \vec{S}\cdot d\vec{\cal A}=-\vec{E}\cdot \frac{\partial \vec{D}}{\partial t}-\vec{H}\cdot \frac{\partial \vec{B}}{\partial t}
$$

Notemos ahora que el segundo miembro no es una diferencial exacta: Esto dá cuenta de que que los campos en los medios materiales no son "sistemas cerrados" sino que intercambian energía con otros grados de libertad, vía las corrientes de polarización. En el caso de medios lineales,
$$
\vec{\cal P}=\chi_E \vec{E}
$$y
$$
\vec{\cal M}=\chi_H \vec{H}
$$
de manera que
$$
\vec{E}\cdot \frac{\partial \vec{D}}{\partial t}=\frac{\partial }{\partial t}(U_E+\frac{1}{2}\vec{\cal P}\cdot\vec{E})
$$
$$
\vec{H}\cdot \frac{\partial \vec{B}}{\partial t}=\frac{\partial }{\partial t}(U_B-\frac{1}{2}\vec{\cal M}\cdot\vec{B})
$$
se reducen a diferenciales exactas: el trabajo sobre las cargas y corrientes de polarización es conservativo (depende sólo del estado de los campos).





## Energía eléctrica y magnética totales

Volvamos ahora a las expresiones para las energías de los campos eléctricos y magnéticos, pero expresándolas ahora en términos de los campos auxiliares $\vec{D}$ y $\vec{H}$:
$$
U_E= \int \frac{\varepsilon_0}{2}|\vec{E}|^2d{\cal V}=
\int \frac{1}{2 \varepsilon_0}|\vec{D}|^2d{\cal V}+
\int \frac{1}{2 \varepsilon_0}|\vec{\cal P}|^2d{\cal V}+
\int \frac{1}{\varepsilon_0}\vec{D}\cdot \vec{\cal P} d{\cal V}
$$
$$
U_B= \int \frac{1}{2 \mu_0}|\vec{B}|^2d{\cal V}=
\int \frac{\mu_0}{2}|\vec{H}|^2d{\cal V}+
\int \frac{\mu_0}{2}|\vec{\cal M}|^2d{\cal V}+
\int \mu_0 \vec{H}\cdot \vec{\cal M} d{\cal V}
$$

En ambos casos, vemos tres contribuciones: la primera, relacionada con los campos "libres", la segunda con el estado de polarización (eléctrica o magnética) de los medios materiales, y un tercer término que involucra la interacción entre los campos y la polarización.   

Notemos que para hacer esta separación no hicimos ninguna hipótesis sobre la relación entre $\vec{D}$ y $\vec{H}$ y las correspondientes polarizaciones ($\vec{\cal P}$ y $\vec{\cal M}$). En el caso lineal, el efecto neto se puede expresar introduciendo una permitividad ($\varepsilon$) y una permeabilidad ($\mu$) efectivas

$$
U_E= 
\int \frac{1}{2 \varepsilon}|\vec{D}|^2d{\cal V}
$$
$$
U_B= 
\int \frac{\mu}{2}|\vec{H}|^2d{\cal V}$$
ya que $\vec{\cal P}$ y $\vec{\cal M}$ ya no son grados de libertad independientes.


**Derivación del teorema de Poynting**

El trabajo realizado por unidad de tiempo por los campos electromagnéticos sobre las corrientes del sistema (incluyendo las de conducción y las de magnetización) viene dado por

\begin{eqnarray}
\frac{\delta W_{total}}{\delta t}&=&\int \vec{E}\cdot\vec{j} dV\\
&=&
\int \vec{E}\cdot(\nabla \times \vec{B}- \frac{\partial \vec{E}}{\partial t}) dV \\
&=&\int (\vec{E}\cdot\nabla \times \vec{B}- \vec{E}\cdot\frac{\partial \vec{E}}{\partial t}) dV\\
&=&
\int (-\nabla \cdot (\vec{E}\times \vec{B})/\mu_0 + \vec{B}\cdot (\nabla \times \vec{E})/\mu_0 - \vec{E}\cdot\frac{\partial \vec{E}}{\partial t}) dV\\
&=&
\int (-\nabla \cdot (\vec{E}\times \vec{B})/\mu_0 - \vec{B}\cdot \frac{\partial \vec{B}}{\partial t}/\mu_0 - \vec{E}\cdot\frac{\partial \vec{E}}{\partial t}) dV\\
\end{eqnarray}

Luego,

\begin{eqnarray}
\frac{\delta W_{EM}}{\delta t} +\int_{\partial} \vec{S}\cdot d\vec{A} =  - \frac{\partial U_E+U_B}{\partial t}=0
\end{eqnarray}
con $\vec{S}=\frac{\vec{E}\times \vec{B}}{\mu_0}$ el *vector de Poynting*: la suma del trabajo neto realizado por los campos y la energía irradiada (flujo del vector de Poynting) es igual a (menos) el cambio en la energía de los campos eléctricomagnéticos.

**Derivación del teorema de Poynting para las corrientes libres**

Si en lugar de considerar a todas las corrientes, queremos evaluar el trabajo de los campos sobre las corrientes libres (de conducción) la derivación anterior se modifica según


\begin{eqnarray}
\frac{\delta W_{libre}}{\delta t}&=&\int \vec{E}\cdot\vec{j}_{libres}dV\\
&=&
\int \vec{E}\cdot(\nabla \times \vec{H}- \frac{\partial \vec{D}}{\partial t}) dV \\
&=&\int (\vec{E}\cdot\nabla \times \vec{H}- \vec{E}\cdot\frac{\partial \vec{D}}{\partial t}) dV\\
&=&
\int (-\nabla \cdot (\vec{E}\times \vec{H}) + \vec{H}\cdot (\nabla \times \vec{E}) - \vec{E}\cdot\frac{\partial \vec{D}}{\partial t}) dV\\
&=&
\int (-\nabla \cdot (\vec{E}\times \vec{H}) - \vec{H}\cdot \frac{\partial \vec{B}}{\partial t} - \vec{E}\cdot\frac{\partial \vec{D}}{\partial t}) dV\\
\end{eqnarray}

Luego,

\begin{eqnarray}
\frac{\delta W_{libre}}{\delta t} + \int_{\partial} \vec{S}\cdot d\vec{A}
= -\int  \vec{H}\cdot \frac{\partial \vec{B}}{\partial t} dV -\int  \vec{E}\cdot \frac{\partial \vec{D}}{\partial t} dV 
\end{eqnarray}

Notamos que ahora el segundo miembro no es en general una diferencial exacta: Si todos los medios son lineales,
\begin{eqnarray}
\frac{\delta W_{libre}}{\delta t} + \int_{\partial} \vec{S}\cdot d\vec{A}
= -\int   \frac{\partial \vec{H}\cdot\vec{B}/2}{\partial t} dV - \frac{\partial \vec{D}\cdot \vec{E}/2}{\partial t}
\end{eqnarray}
y tiene sentido definir densidades de energía Eléctrica y Magnética "efectivas".
$$
U_D = \int \frac{\vec{D}\cdot \vec{E}}{2}dV=\int \varepsilon_0 \frac{\vec{E}\cdot \vec{E}}{2}dV-\int \frac{\vec{\cal P}\cdot \vec{E}}{2}dV
$$

$$
U_H = \int \frac{\vec{H}\cdot \vec{B}}{2}dV=\int \frac{\vec{B}\cdot \vec{B}}{2\mu_0}dV-\int \frac{\vec{\cal M}\cdot \vec{B}}{2}dV
$$
Por otro lado, en el caso de medios no lineales, esto no es posible. Esto significa que en general, si los campos externos se hacen evolucionar periódicamente (se ciclan) la integral del segundo miembro no se anula en un ciclo. Despreciando las pérdidas por radiación, el trabajo neto de los campos es no nulo, y da cuenta de las pérdidas energéticas producidas en el material (análogo al efecto Joule).


# Disipación  en medios con histéresis magnética


El fenómeno de histéresis magnética se dá cuando la magnetización en un medio material no depende sólo de la configuración actual de los campos magnéticos, sino también de su historia.  Podemos entender esto partiendo de un modelo microscópico simple. Supongamos que la magnetización del material proviene de la contribución de una densidad de momentos magnéticos microscópicos, que son libres de orientarse entre sí. Si inicialmente se encuentran todos orientados al azar, el campo magnético neto que generan estos dipolos dentro de la muestra será nulo.  Por otro lado, en presencia de un campo magnético externo, los momentos magnéticos tratarán de  alinearse en la dirección del campo, pasando a una fase "ordenada". En esa fase, el campo neto generado por los dipolos ya no es cero, sino que tomará un valor finito, orientándose en el mismo sentido que el campo externo.
Si ahora se desconecta el campo externo, como los momentos siguen alineados, continuarán sosteniendo un campo magnético en la dirección que tenía el campo externo, que hará energéticamente desfavorable los cambios de orientación de los momentos individuales.

Si en lugar de apagar repentinamente el campo, se lo hace variar "lentamente"  la magnetización del sistema mantendrá una relación funcional con este. Esta relación funcional es tal que al anularse el campo externo, se alcanza una magnetización no nula. La relación funcional se mantendrá en tanto no se alcance un cierto valor de saturación. Cuando esto ocurre, la relación funcional cambia, de forma que la magnetización remanente cambia de signo.

Un modelo típico para estas relaciones funcionales es de la forma
$$
{\cal M}(H)=\frac{{\cal M}_s}{\pi/2}\arctan(\frac{\pi \chi}{2}\frac{H\pm H_c}{{\cal M}_S} )
$$

donde el signo delante de $H_c$ dependerá del sentido de $H$ la última vez que se alcanzó la saturación ($|H|\gg \frac{2}{\chi\pi}{\cal M}_s$)



Consideremos ahora una muestra con este comportamiento, en el interior de un solenoide, alimentado por una corriente $i$, que produce un campo $H= \mu_0 n i$ en la dirección de su eje de simetría.

Si la corriente varía de manera armónica con frecuencia $\omega/2\pi$, $H$ hará lo mismo
$$
H(t)=H_0 \cos(\omega t)
$$
Supongamos que $H_0$ es del orden del campo de saturación. Luego, el momento magnético de la muestra, como función del tiempo, será de la forma

$$
M(t)= \frac{M_0}{\pi/2}\arctan\left(  \frac{\pi \chi }{2{\cal M}_S} (H_0 \cos(\omega t) + H_c \frac{\sin(\omega t)}{|\sin(\omega t)|})  \right) 
$$


con $M_0={\cal M}_s {\cal V}$ el momento magnético total de la muestra. La potencia eléctromagnética absorbida por el material se expresa entonces como

$$
W=-\int H\frac{d B}{dt} d{\cal V}=-\frac{d }{dt} \frac{H^2 {\cal V}}{2\mu_0 }  -
\frac{H}{\mu_0}\frac{d M}{dt}
$$

El primer término es la contribución a la energía magnética del solenoide en el volumen ocupado por la muestra, y estaría ahí aunque no hubiese muestra. El segundo término representa entonces el verdadero efecto de la muestra. Recordemos que debido a su momento magnético, la muestra modifica el flujo magnético en el interior del solenoide, lo que induce una FEM en el solenoide. Esta FEM debe ser compensada por la fuente de alimentación que impulsa la corriente, que "ve" a la muestra como si fuese una resistencia extra en el conductor del solenoide.

Integrando esta cantidad en medio ciclo (donde $M(t)$ es continua), 
$$
W_1=\int_0^{\pi/\omega} \frac{M}{\mu_0}\frac{d H}{dt}dt=\frac{M_0}{\pi/2 \mu_0} \int_{-H_0}^{H_0} \arctan\left(  \frac{\pi \chi }{2{\cal M}_S} (H+ H_c)  \right) dH=
-\frac{4 {\cal M}_s M_0}{\pi^2 \mu_0 \chi} \int_{\frac{\pi \chi }{2{\cal M}_S} (-H_0+ H_c)}^{\frac{\pi \chi }{2{\cal M}_S} (H_0+ H_c)} \arctan\left(  u  \right) du
$$

y en el siguiente  medio ciclo, 
$$
W_2=\int_0^{\pi/\omega} \frac{M}{\mu_0}\frac{d H}{dt}dt=\frac{M_0}{\pi/2 \mu_0} \int_{-H_0}^{H_0} \arctan\left(  \frac{\pi \chi }{2{\cal M}_S} (H+ H_c)  \right) dH=
-\frac{4 {\cal M}_s M_0}{\pi^2 \mu_0 \chi} \int_{\frac{\pi \chi }{2{\cal M}_S} (-H_0- H_c)}^{\frac{\pi \chi }{2{\cal M}_S} (H_0- H_c)} \arctan\left(  u  \right) du=-W_1
$$

La contribución neta al trabajo viene entonces de los puntos en que cuando $M(t)$ es discontinua: en esos puntos,

\begin{eqnarray}
W_{disc,\pm}&=& -\int_{T-\epsilon}^{T+\epsilon} H \frac{dM}{dt} dt  \\
&=& \pm H_0 (M(T-\epsilon)-M(T+\epsilon))\\
&=&
\frac{\pm  H_0 M_0}{\pi/2}(\arctan( \frac{ \pi\chi}{2} \frac{\mp H_0\pm H_c}{M_S})-\arctan( \frac{ \pi\chi}{2} \frac{\mp H_0\mp H_c}{M_S}))\\
&=&\frac{H_0 M_0}{\pi/2}(\arctan(- \frac{ \pi\chi}{2} \frac{H_0- H_c}{M_S})+\arctan( \frac{ \pi\chi}{2} \frac{ H_0+ H_c}{M_S}))\\
&\approx& \frac{4  M_0 }{\pi^2 \chi}\frac{1}{H_0/H_c-H_c/H_0}
\end{eqnarray}


Finalmente, la potencia media disipada resulta ser


\begin{eqnarray}
P&=& \frac{\omega}{2\pi} 2 W_{disc,\pm}
\approx \frac{4  M_0  \omega}{\pi \chi}\frac{H_c}{H_0-H^2_c/H_0}
\end{eqnarray}




# Graficas

En el código siguiente, podemos explorar la evolución de la magnetización y los intercambios de energía para el modelo planteado. Como ejercicio, observe qué ocurre al variar los parámetros $h0$ y $hc$

```python
from __future__ import print_function, division
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# Parámetros
h0 = 10
hc = 1

h_old = 0
def mag(h):
    global h_old
    if type(h) is np.ndarray:
        h_old = 0
        return np.array([mag(hh) for hh in h])
    
    if h_old < h:
        hcs = 1
    else:
        hcs = -1
    h_old = h
    return np.arctan(h+hc*hcs)


def work(h):
    global h_old
    if type(h) is np.ndarray:
        h_old = 0
        return np.array([work(hh) for hh in h])
    
    if h_old < h:
        hc = 1
    else:
        hc = -1
    dm = np.arctan(h+hc)-np.arctan(h_old+hc)
    db = dm + h-h_old
    w = h*db
    h_old = h
    return w
  
    
    
    
hs = h0 * np.cos(np.linspace(0.1,4*np.pi,100))
ms =  mag(hs)
bs = hs+ ms

########################################3
fig,axs = plt.subplots(1,2,figsize=(10,5))
ax = axs[0]
ax.plot(hs,ms,label="M")
ax.set_xlabel("h")
ax.set_ylabel("m")
ax.set_title("h vs m")
ax.legend()
ax=axs[1]
ax.plot(hs,bs,label="b")
ax.set_xlabel("h")
ax.set_ylabel("b")
ax.set_title("h vs b")
ax.legend()
plt.show()
##########################

```

```python

ubs = bs*bs/2.
uhs = hs*hs/2.
plt.plot(uhs,label="$U_{H}$")
plt.plot(ubs,label="$B_{H}$")
plt.title("U magnética libre vs total")
plt.legend()
plt.show()

```

```python

fig,axs = plt.subplots(1,2,figsize=(10,5))
ax = axs[0]

ws = work(hs)
ax.plot(ws[1:],label="Potencia")
ax.plot(.1*hs[1:],label="h")
ax.plot(ms[1:],label="m")
ax.set_title("Potencia absorbida vs t")
ax.set_xlabel("t")
ax.set_ylabel("P")
ax.legend()
ax = axs[1]
ax.plot(np.array([sum(ws[0:i]) for i in range(1,len(ws))]),label="trabajo em")
ax.set_xlabel("t")
ax.set_ylabel("W")
ax.set_title("Trabajo consumido vs t")
ax.legend()
plt.show()


```
